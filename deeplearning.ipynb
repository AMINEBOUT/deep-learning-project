```python
from keras.datasets import mnist
import numpy as np
import keras
from keras import layers
```

```python
#? Loading the training/test data
(train_images, train_lables), (test_images, test_lables) = mnist.load_data()
```

```python
# the models mods
model = keras.Sequential(
    [
        layers.Dense(512, activation="relu"),
        layers.Dense(10, activation="softmax"),
    ]
)
```

```python
#? compile types
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"],
)
```

```python
# adjust the shape
#! need to reshape the test & train images
train_images = train_images.reshape(60000, 28*28)
tarin_images = train_images.astype("float32")/255

test_images = test_images.reshape(10000, 28*28)
test_images = test_images.astype("float32")/255
```

```python
#? training the model
model.fit(train_images, train_lables, epochs=6, batch_size=126)
```

```python
#? check the results

test_digits = test_images[0:10]
prediction = model.predict(test_digits)
print("Prediction:", prediction[9].argmax())
print("Real result:", test_lables[9])
```

```python
#? check accuracy and loss
test_loss, test_acc = model.evaluate(test_images, test_lables)
```
